```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Year": [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "Population": [2.53, 3.03, 3.71, 4.45, 5.32, 6.12, 6.92, 7.79]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Population"], marker='o', linestyle='-', linewidth=2)
plt.xlabel("Year")
plt.ylabel("World Population (Billions)")
plt.title("World Population Growth Over Decades")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Date": pd.date_range(start="2023-01-01", periods=10, freq="M"),
    "Close": [129, 135, 140, 145, 138, 150, 155, 160, 158, 165]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], marker='o', linestyle='-', linewidth=2)
plt.xlabel("Date")
plt.ylabel("AAPL Close Price (USD)")
plt.title("Apple (AAPL) Stock Price Trend")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
